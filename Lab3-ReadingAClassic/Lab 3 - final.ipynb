```python
#Fatma Betul Seker
#Lab 3 - Reading a Classic
#3 Feb 2022
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
frank_book = open("84-0.txt", "r", encoding="utf8")
fbook = frank_book.read().replace("\n"," ").replace("Henry Clerval","Henry").replace("Victor Frankenstein","Victor").replace(",", " ").replace(".", " ")
#Replacing "Henry Clerval" with "Henry" so I can count it as one if they appear together
frank_book.close()
```

```python
fbook_list = fbook.split(" ")
char_list = ["Agatha", "Caroline","Clerval","creature","Elizabeth","Ernest","Felix","Frankenstein","Henry","Justine", "Lacey","Victor", "William"]
```

```python
print(fbook_list.index("Saville"))
#I'm going to start finding the indexes of the chapters from index 301 to avoid counting the Intro and CONTENTS page
```

```python
chp_no=1
#Finding the initial index for Chapter 1
primary_index=fbook_list.index("Chapter",301)
#Adding Chapter at the end of the list so I can scan "between" chapters
fbook_list.append("Chapter")
while chp_no <= 24:
    print("         ")
    print("Chapter",chp_no)
    print("------------------------------")
    secondary_index=fbook_list.index("Chapter", primary_index+1)
    chp_list=fbook_list[primary_index:secondary_index]
    chp_no=chp_no+1
    primary_index = secondary_index
    for char in char_list:
        print(char,chp_list.count(char))
    
```
