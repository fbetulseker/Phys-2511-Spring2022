```python
#Fatma Betul Seker
#Lab 4 - Blindfolded Archer
#10 Feb 2022

import numpy as np
import math as ma
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
#Constant values for the calculations

#Archer's shoulder height
y_0 = 1.7 #m

#Gravitational acc.
g = 9.8 #m/s^2

#Air drag coefficient for arrow
b = 0.04

#Time intervals
dt = 0.01 #s
```

```python
hit = False

while hit == False: 
    v_0 = np.random.randint(9,40) #m/s
    theta_0 = np.random.rand() * ma.pi / 4 #radian
    
    #Arrays for horizontal motion
    x = np.array([0])
    v_x = np.array([v_0*ma.cos(theta_0)])
    a_x = np.array([-v_0*ma.cos(theta_0)*b])
    
    #Arrays for vertical motion
    y = np.array([y_0])
    v_y = np.array([v_0*ma.sin(theta_0)])
    a_y = np.array([-g -v_0*ma.sin(theta_0)*b])
    
    
    i = 0
    while y[i] > 0:
        y = np.append(y, dt*v_y[i] + y[i])
        x = np.append(x, dt*v_x[i] + x[i])

        v_y = np.append(v_y, dt*a_y[i] + v_y[i])
        v_x = np.append(v_x, dt*a_x[i] + v_x[i])

        i = i+1

        a_y = np.append(a_y, -g -v_y[i]*b)
        a_x = np.append(a_x, -v_x[i]*b)
    
    #Creating target on the plot--------------------------------------------------
    fig, ax = plt.subplots()
    ax.plot(x, y)
    target = Rectangle((10, 0), 2, 0.01, linewidth=1, edgecolor='r', facecolor='red')
    ax.add_patch(target)
    plt.show()
    #Credit: matplotlib.org/stable/gallery/statistics/errorbars_and_boxes
    
    #Target as an array
    target_array = np.array([10,11,12])
    #Target's midpoint: 11
    #Target's lenght: 2

    
    x_final = x[i]
    
    if abs(11 - x_final) < 1:
        print("Congratulations!! You hit the target")
        hit = True

    else:
        print("Oops! You missed the target by", abs(11 - x_final), "meters")
    
    
    #Arrays for horizontal motion
    x = np.array([0])
    v_x = np.array([v_0*ma.cos(theta_0)])
    a_x = np.array([0])
    
    #Arrays for vertical motion
    y = np.array([y_0])
    v_y = np.array([v_0*ma.sin(theta_0)])
    a_y = np.array([-g])
    
    i = 0
    while y[i] > 0:
        y = np.append(y, dt*v_y[i] + y[i])
        x = np.append(x, dt*v_x[i] + x[i])

        v_y = np.append(v_y, dt*a_y[i] + v_y[i])
        v_x = np.append(v_x, dt*a_x[i] + v_x[i])

        i = i+1

        a_y = np.append(a_y, -g)
        a_x = np.append(a_x, 0)
    
    x_final_no = x[i]
    
    print("You have been impacted by the air resistance by", abs(x_final - x_final_no), "meters")
    

```
